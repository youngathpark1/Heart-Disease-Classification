## Preprocessing: Dummy Variables
As part of data preprocessing, I create dummy variables for the features:

- sex (already binarized)
- cp
- fbs (already binarized)
- restecg
- exang (already binarized)
- slope
- ca
- thal

```python
# Imports

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the data

df = pd.read_csv('./heart.csv')

df.head()
```

```python
# Using pandas get_dummies, I dummify the following features

df = pd.get_dummies(data=df, columns=['cp', 'restecg', 'slope', 'ca', 'thal'], drop_first=True)

df.head()
```

```python
# Exporting modified df as csv to reference during modeling

df.to_csv('./heart_modified.csv', index=False)
```

## Conclusions & Recommendations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
model_performance = {
    'Model' : ['Log', 'KNN', 'RF', 'SVM', 'XG'],
    'Training_Acc' : [0.8792682926829268, 1.0, 0.9365853658536586, 0.998780487804878, 1.0],
    'Testing_Acc' : [0.8634146341463415, 1.0, 0.9121951219512195, 0.9853658536585366, 1.0]
}
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(data=model_performance, x='Training_Acc', y='Testing_Acc', hue='Model', s=150)
plt.xlabel(xlabel='Training Accuracy', fontsize=15)
plt.ylabel(ylabel='Testing Accuracy', fontsize=15)
plt.title('Training/Testing Accuracy Scores', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12);
```

KNN and XGBoost performed significantly well with this data set. In terms of accuracy, both KNN and XGBoost scored 100% accuracy score in both training and testing data. Logistic Regression performed the worst with training and testing accuracy scores in the high 80s while other models hovered in the 90s.

Both KNN and XGBoost also scored 100% recall score which essentially means it did not incorrectly predict positives as negatives which is a fact worth highlighting since false negatives can have a big consequences in this particular data scenario where the goal is to predict whether a patient has a heart disease or not. In fact, in any medical-related studies, it is critical that Type II errors are weighted heavily since the implications can be magnified in a medical/research related fields.

While KNN and XGBoost performed very well, one downside to consider is that there is little to no room for any interpretability. If interpretability is a major factor of consideration, Logistic Regression may be of better use since coefficients are interpretable.

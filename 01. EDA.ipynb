```python
# Function that creates histograms

def create_hist(df):
    df = pd.read_csv('./heart.csv')
    cols = df.columns
    
    for col in cols:
        plt.figure(figsize=(15,8))
        sns.histplot(df[col])
        plt.title(f'Distribution of {col}')
```

```python
create_hist(df)
```
